```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
df=pd.read_csv("Churn_Modelling.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace=True)
```

```python
le=LabelEncoder()
df['Geography']=le.fit_transform(df['Geography'])
df['Gender']=le.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
x=df.drop('Exited',axis=1)
y=df['Exited']
```

```python
x.shape
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print("Training classes(x):",x_train.shape)
print("Training classes(x):",y_train.shape)
print("Training classes(x):",x_test.shape)
print("Training classes(x):",y_test.shape)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU,Dropout
```

```python
classifier=Sequential()
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=10))
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['Accuracy'])
```

```python
model_training=classifier.fit(x_train,y_train,batch_size=10,epochs=100)
```

```python
classifier.summary()
```

```python
predictions=classifier.predict(x_test)
predictions=(predictions>0.5)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,predictions)
score
```
